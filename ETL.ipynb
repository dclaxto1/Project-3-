```python
# Dependencies
import requests
import json
import pandas as pd
import numpy as np
import sqlite3
```

```python
# URL for GET requests to retrieve weather data
og_url = "https://archive-api.open-meteo.com/v1/archive?latitude=33.749&longitude=-84.388&start_date=2022-01-01&end_date=2022-12-31&daily=weathercode,temperature_2m_max,precipitation_sum,rain_sum,snowfall_sum,precipitation_hours,windspeed_10m_max&temperature_unit=fahrenheit&windspeed_unit=mph&precipitation_unit=inch&timezone=America%2FChicago"

cities = {
    "atlanta": [33.749,-84.388],
    "philly": [39.952583,-75.165222],
    "chicago": [41.881832,-87.623177]
}
```

```python
#using loop to pull all weather info the cities in our city dictionary
city_weather = {}
for city, coordinates in cities.items():
    lat = coordinates[0]
    lon = coordinates[1]
    url = (f"https://archive-api.open-meteo.com/v1/archive?latitude={lat}&longitude={lon}&start_date=2022-01-01&end_date=2022-12-31&daily=weathercode,temperature_2m_max,precipitation_sum,rain_sum,snowfall_sum,precipitation_hours,windspeed_10m_max&temperature_unit=fahrenheit&windspeed_unit=mph&precipitation_unit=inch&timezone=America%2FChicago")
    weather = requests.get(url).json()
    city_weather[city]=weather
    print(f"city: {city}, lat: {lat}, lon: {lon}")
    print(f"API URL {city}:", url)
    
```

```python
#pulling Atlanta values from API call for our new dataframe
atlanta_data = {
    "city_id": [1] * 365,
    "date": city_weather['atlanta']['daily']['time'],
    "weathercode": city_weather['atlanta']['daily']['weathercode'],
    "max_temp": city_weather['atlanta']['daily']['temperature_2m_max'],
    "precip_sum": city_weather['atlanta']['daily']['precipitation_sum'],
    "rain_sum": city_weather['atlanta']['daily']['rain_sum'],
    "snow_sum": city_weather['atlanta']['daily']['snowfall_sum'],
    "precip_hours": city_weather['atlanta']['daily']['precipitation_hours'],
    "max_windspeed": city_weather['atlanta']['daily']['windspeed_10m_max']
}
```

```python
#pulling Philly values from API call for our new dataframe
philly_data = {
    "city_id": [2] *365,
    "date":city_weather['philly']['daily']['time'],
    "weathercode": city_weather['philly']['daily']['weathercode'],
    "max_temp": city_weather['philly']['daily']['temperature_2m_max'],
    "precip_sum": city_weather['philly']['daily']['precipitation_sum'],
    "rain_sum": city_weather['philly']['daily']['rain_sum'],
    "snow_sum": city_weather['philly']['daily']['snowfall_sum'],
    "precip_hours": city_weather['philly']['daily']['precipitation_hours'],
    "max_windspeed": city_weather['philly']['daily']['windspeed_10m_max']
}
```

```python
#pulling Chicago values from API call for our new dataframe
chicago_data = {
    "city_id": [3] * 365,
    "date": city_weather['chicago']['daily']['time'],
    "weathercode": city_weather['chicago']['daily']['weathercode'],
    "max_temp": city_weather['chicago']['daily']['temperature_2m_max'],
    "precip_sum": city_weather['chicago']['daily']['precipitation_sum'],
    "rain_sum": city_weather['chicago']['daily']['rain_sum'],
    "snow_sum": city_weather['chicago']['daily']['snowfall_sum'],
    "precip_hours": city_weather['chicago']['daily']['precipitation_hours'],
    "max_windspeed": city_weather['chicago']['daily']['windspeed_10m_max']
}
```

```python
#Create Atlanta, Philly, and Chicago Dataframes
atlanta_weather = pd.DataFrame(atlanta_data)
philly_weather = pd.DataFrame(philly_data)
chicago_weather = pd.DataFrame(chicago_data)
```

```python
#changing 'date' to date format for all dataframes
atlanta_weather["date"] = pd.to_datetime(atlanta_weather["date"])
philly_weather["date"] = pd.to_datetime(philly_weather["date"])
chicago_weather["date"] = pd.to_datetime(chicago_weather["date"])
```

```python
#export dataframe to sqlite file
filename = 'city_weather_info.db'
conn = sqlite3.connect(filename)
cur = conn.cursor()
```

```python
#creating new tables in our sqlite datebase file
atlanta_weather.to_sql('atl_weather',conn, if_exists='replace',index=False)
philly_weather.to_sql('philly_weather',conn, if_exists='replace',index=False)
chicago_weather.to_sql('chicago_weather',conn, if_exists='replace',index=False)
```

```python
#compressing tables into one table
cur.execute("CREATE TABLE if not exists combined_weather AS SELECT * FROM atl_weather UNION SELECT * FROM philly_weather UNION SELECT * from chicago_weather")
for row in cur:
    print(row)
```

```python
#displaying first 10 records of new table
cur.execute("select * from combined_weather where city_id = 3 limit 10")
for row in cur:
    print(row)
```

```python
#creating fact table
cur.execute('''CREATE TABLE if not exists city_list (
           city_id INTEGER PRIMARY KEY,
            city_name TEXT,
            lat FLOAT,
            lon FLOAT)''')
conn.commit()
```

```python
#Inserting data into new table
new_rows = [(1, 'Atlanta', 33.7490, -84.3880),
            (2, 'Philadelphia', 39.9526, -75.1652),
            (3, 'Chicago', 41.881832, -87.623177)]
cur.executemany("INSERT INTO city_list (city_id, city_name, lat, lon) VALUES (?, ?, ?, ?)", new_rows)
conn.commit()
```
